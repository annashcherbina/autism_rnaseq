```python
gene_list=['PARP1','CASP3','CD24','CD247','B2M']
```

```python
import pandas as pd 
de_dir="/Users/annashch/autism_rnaseq/post_processing"
from os.path import join 

de_asddm_asdn=pd.read_csv(join(de_dir,"ConditionASDDM_vs_ConditionASDN.de.tsv"),header=0,sep='\t')
de_asddm_tdn=pd.read_csv(join(de_dir,"ConditionASDDM_vs_ConditionTDN.de.tsv"),header=0,sep='\t')
de_asdn_tdn=pd.read_csv(join(de_dir,"ConditionASDN_vs_ConditionTDN.de.tsv"),header=0,sep='\t')
```

```python
de_asddm_asdn.head()
```

```python
g1=de_asddm_asdn.loc[gene_list]
g1['comparison']='ASDDM_vs_ASDN'
g1.head()
```

```python
g2=de_asddm_tdn.loc[gene_list]
g2['comparison']='ASDDM_TDN'
g2.head()
```

```python
g3=de_asdn_tdn.loc[gene_list]
g3['comparison']='ASDN_TDN'
g3.head()
```

```python
pd.concat([g1,g2,g3]).sort_index().to_csv('genes_of_interest.txt',header=True,index=True,sep='\t')
```
