```python
import matplotlib 
import pandas as pd
import numpy as np 

from matplotlib import pyplot as plt
def qqplot(pv,title):
    nlpvo = -np.log10(np.sort(pv))
    nlpve = -np.log10(np.linspace(0,1,pv.shape[0]+2)[1:-1])
    plt.figure()
    plt.plot(nlpve, nlpvo, '.k')
    plt.plot([0, 10], [0, 10], 'k-', color = 'r')
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.xlabel("Expected -log10(p)") 
    plt.ylabel("Observed -log10(p)") 
    plt.title(title)
```

```python
#get all the .de.tsv files 
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir('permuted') if isfile(join('permuted', f))]
```

```python
for f in onlyfiles: 
    data=pd.read_csv(join('permuted',f),header=0,sep='\t')
    qqplot(data['P.Value'],f)
```
